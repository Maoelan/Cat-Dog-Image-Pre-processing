```python
#Memasukkan library yang dibutuhkan
from os import listdir
import tensorflow as tf
import keras
import cv2
import matplotlib.pyplot as plt
import numpy as np
from keras.models import load_model
import keras.layers as L
```

```python
#Menampilkan citra asli
plt.figure(figsize=(9,6))
for id, img in enumerate(listdir('../content/training_set/training_set/cats/kucing')[0:3]):
    img = cv2.imread('../content/training_set/training_set/cats/kucing/'+img)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.subplot(2,3,id+1)
    plt.imshow(img)
for id, img in enumerate(listdir('../content/training_set/training_set/dogs/anjing')[0:3]):
    img = cv2.imread('../content/training_set/training_set/dogs/anjing/'+img)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.subplot(2,3,id+4)
    plt.imshow(img)
```

```python
#Mengubah citra ke grayscale
plt.figure(figsize=(9,6))
for id, img in enumerate(listdir('../content/training_set/training_set/cats/kucing')[0:3]):
    img = cv2.imread('../content/training_set/training_set/cats/kucing/'+img)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.subplot(2,3,id+1)
    plt.imshow(img)
for id, img in enumerate(listdir('../content/training_set/training_set/dogs/anjing')[0:3]):
    img = cv2.imread('../content/training_set/training_set/dogs/anjing/'+img)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.subplot(2,3,id+4)
    plt.imshow(img)
```

```python
len_train_set = len(listdir('../content/training_set/training_set/dogs/anjing'))+len(listdir('../content/training_set/training_set/cats/kucing/'))
len_test_set = len(listdir('../content/training_set/training_set/dogs/anjing'))+len(listdir('../content/training_set/training_set/cats/kucing/'))
```
